##  What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
df[(df.horsepower>100) & (df.weight<3000)].origin
```

```python
df[(df.horsepower>100) & (df.weight<3000)].origin.mode()[0]
```

## What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
df[df.origin=='japan'].acceleration.mean().round(2)
```

##  Which year had the highest average mpg?

```python
df.groupby('model_year')['mpg'].mean()
```

## Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
max_metric=df[ df.mpg>df.mpg.median()].metric.max()
```

```python
df[ (df.mpg>df.mpg.median()) & (df.metric==max_metric)]
```

## X = horsepower

## Y = weight

## Color by: origin

## Size by: mpg

## Hue order = ['japan', 'europe', 'usa']

## Add meaningful plot titles and axis titles.

```python
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200)
)

plt.title('weight vs horsepower')
plt.legend() 
plt.show()
```
